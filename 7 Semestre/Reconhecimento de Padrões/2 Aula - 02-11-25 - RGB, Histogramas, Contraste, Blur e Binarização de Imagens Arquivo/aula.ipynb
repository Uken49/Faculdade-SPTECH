# Tratando a imagem para ficar em tons de cinza

```python
from matplotlib import pyplot as plt
import cv2

img = cv2.imread('carro.jpg', cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Imagem", img)

hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.figure()
plt.title("Histograma em Tons de Cinza")
plt.xlabel("Intensidade")
plt.ylabel("Qtde. de Pixels")
plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

```python
img = cv2.imread('/content/samu.png', cv2.IMREAD_UNCHANGED)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')

color = ("r", "g", "b")

plt.figure(figsize=(10, 5))
plt.title("Histograma em Tons de Cinza")
plt.xlabel("Intensidade")
plt.ylabel("Qtde. de Pixels")

for i, cor in enumerate(color):
  hist = cv2.calcHist([img], [0], None, [256], [0, 256])
  plt.plot(hist, color=cor)
  plt.xlim([0,256])

plt.show()
```
